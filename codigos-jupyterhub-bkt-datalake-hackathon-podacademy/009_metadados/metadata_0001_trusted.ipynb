```pyspark
meta_join_00 = spark.sql(     
    """
        select
            a.zone,
            a.table,
            lower(a.column) as column,
            a.type,
            --b.table as table_b,
            --b.column as column_b,
            --b.type_object, 
            b.description
        from
            metadata_structure as a
            left join meta_desc as b
        on (a.table = b.table and lower(a.column) = lower(b.column))
        
    """)
meta_join_00.registerTempTable("meta_join_00")
meta_join_00.cache()
meta_join_00.count()
```
