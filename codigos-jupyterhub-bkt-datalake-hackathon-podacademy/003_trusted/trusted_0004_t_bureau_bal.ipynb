```pyspark
import os
import sys
import pytz
import numpy as np
import datetime
from pyspark.sql import SparkSession
from pyspark.sql import SQLContext
from pyspark.sql.functions import udf, lpad, translate
from datetime import datetime 
from datetime import timedelta
from datetime import date
from dateutil.relativedelta import relativedelta
from pyspark.sql.types import *
from pyspark.sql.functions import count, avg
```

```pyspark
agora=datetime.now(pytz.timezone('America/Sao_Paulo'))
dthproc=agora.strftime("%Y%m%d%H%M%S")
```

```pyspark
spark = SparkSession.builder.appName('trusted_bureau_balance_20230805').getOrCreate()
sqlContext=SQLContext(spark.sparkContext)
```

```pyspark
_DIA_AULA_ = '20230804'
_REF_ = '202001'
```

```pyspark
raw ='t_bureau_balance_20230805'
```

```pyspark
bucket_trusted = 'bkt-datalake-grupo-01-hackathon-podacademy/{dia}/003_trusted'.format(dia=_DIA_AULA_)
bucket_raw = 'bkt-datalake-grupo-01-hackathon-podacademy/{dia}/002_raw'.format(dia=_DIA_AULA_)
bucket_control = 'bkt-datalake-grupo-01-hackathon-podacademy/{dia}/005_control'.format(dia=_DIA_AULA_)
```

```pyspark
output_trusted = 'trusted_bureau_balance_20230805'
```

```pyspark
full_path_raw = 's3://{bkt}/{file}'.format(bkt=bucket_raw,file=raw)
print(full_path_raw)
```

```pyspark
path_lake = os.path.join(full_path_raw)
raw_00=sqlContext.read.format("parquet").load(path_lake)
raw_00.registerTempTable("raw_00")
```

```pyspark
raw_00.show(5,truncate=False)
```

```pyspark
# +------+------------------+--------------+------------+--------------+------+-------------+----------------------------+-----------------+
# |ref   |ts_file_generation|ts_proc       |SK_ID_BUREAU|MONTHS_BALANCE|STATUS|ref_partition|ts_file_generation_partition|ts_proc_partition|
# +------+------------------+--------------+------------+--------------+------+-------------+----------------------------+-----------------+
# |202308|2023080500        |20230806155321|5715448     |0             |C     |202308       |2023080500                  |20230806155321   |
# |202308|2023080500        |20230806155321|5715448     |-1            |C     |202308       |2023080500                  |20230806155321   |
# |202308|2023080500        |20230806155321|5715448     |-2            |C     |202308       |2023080500                  |20230806155321   |
# |202308|2023080500        |20230806155321|5715448     |-3            |C     |202308       |2023080500                  |20230806155321   |
# |202308|2023080500        |20230806155321|5715448     |-4            |C     |202308       |2023080500                  |20230806155321   |
# +------+------------------+--------------+------------+--------------+------+-------------+----------------------------+-----------------+
# only showing top 5 rows
```

```pyspark
print('lista de colunas para tipar')
for col in raw_00.columns:
    print('cast(' + col + ' as) as ' + col + ',')
```

```pyspark
lake = spark.sql(     
    """
        select
            -- padrao para todas as cargas no datalake--
            
            ref,
            ref_partition,
            
            {pdthproc} as ts_proc,
            {pdthproc} as ts_proc_partition,
        
            -- campos do arquivo --

            cast(SK_ID_BUREAU as int) as SK_ID_BUREAU,
            cast(MONTHS_BALANCE as int) as MONTHS_BALANCE,
            cast(STATUS as string) as STATUS
            
        from
            raw_00
        --where
            --ref_partition = {_ref_}
        --    ref = {_ref_}
            
            
    """.format(pdthproc=dthproc,_ref_ = _REF_))
lake.registerTempTable("lake")
lake.cache()
lake.count()  
```

```pyspark
lake.explain()
```

```pyspark
lake.printSchema()
```

```pyspark
lake.show(4,truncate=False)
```

```pyspark
path_trusted = os.path.join('s3://',bucket_trusted,output_trusted)
print(path_trusted)

lake.coalesce(1).write. \
partitionBy("ref_partition","ts_proc_partition"). \
parquet(path_trusted,mode='overwrite')
```

```pyspark
varDataFile = output_trusted
```

```pyspark
controle = spark.sql(    
        """
            select
                '{tb}' as name_file,
                '{tb}' as name_file_partition,
                
                ref,
                ref_partition,
                
                ts_proc,
                ts_proc_partition,
                
                count(*) as qtd_registros
            from 
                lake as a
            group by
                1,2,3,4,5,6
            order by
                1,2,3,4,5,6
        """.format(tb=varDataFile))
```

```pyspark
controle.registerTempTable("controle")
controle.cache()
qtd=controle.count()
```

```pyspark
controle.show(qtd,truncate=False)
```

```pyspark
path_control = os.path.join('s3://',bucket_control,'tb_0002_controle_procesamento_trusted')
print(path_control)
controle.coalesce(1).write.\
parquet(path_control,mode='append')
```
